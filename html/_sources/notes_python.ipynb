### autres exemples de pattern:
    pattern = r"[A-Z][A-Z][0-9]"
--> matches any two uppercase letters followed by a number

    pattern = r[^A-Z]
--> excludes uppercases letters

    pattern = r"(spam)*egg"
--> any "spam" followed by "egg"

    pattern = r"colo(u)?r"
--> matches color or colour

    pattern = r"9{1,3}$"
--> any string with 1 to 3 nines

    pattern = r"gr(a|e)y"   
--> grey or gray



## Rappels RANGE
Start 0, increments 1, stop **before**
- `[:7]` : starts to 7 (not included)
- `[7:]` : 7th to end
- `[::2]` : du début à la fin de 2 en 2
- `[0:-1]` : exclure le dernier (si -2, les deux derniers...
- `[::-1]` : backwards
